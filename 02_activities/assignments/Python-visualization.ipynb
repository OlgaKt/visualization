```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Users/olgaktytor/Downloads/crime-rates.csv')
df
```

```python
data = {
    'Crime Type': ['ASSAULT', 'AUTOTHEFT', 'BIKETHEFT', 'BREAKENTER', 'HOMICIDE', 'ROBBERY', 'SHOOTING', 'THEFTFROMMV', 'THEFTOVER'],
    'Number of Incidents': [
        df['ASSAULT_2023'].sum(),
        df['AUTOTHEFT_2023'].sum(),
        df['BIKETHEFT_2023'].sum(),
        df['BREAKENTER_2023'].sum(),
        df['HOMICIDE_2023'].sum(),
        df['ROBBERY_2023'].sum(),
        df['SHOOTING_2023'].sum(),
        df['THEFTFROMMV_2023'].sum(),
        df['THEFTOVER_2023'].sum()
    ]
}

crime_data = pd.DataFrame(data)

plt.figure(figsize=(10, 7))
sns.barplot(x='Crime Type', y='Number of Incidents', data=crime_data, color = 'navy')
plt.title('Number of Incidents by Crime Type in 2023', fontsize=16)
plt.xlabel('Crime Type', fontsize=16)
plt.ylabel('Number of Incidents', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

**Summary**

The visualization presents the number of incidents of different crime types for year 2023 in Toronto. Crime types include Assault, Auto Theft, Bike Theft, Break and Enter, Homocide, Robbery, Shooting, Theft from Vehicles, and Theft Over.

Data Source: https://open.toronto.ca/dataset/neighbourhood-crime-rates/
